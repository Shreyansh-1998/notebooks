```python
import pandas as pd
```

```python
#importing the dataset
messages=pd.read_csv('NLP/smsspamcollection.txt', sep='\t',names=["label","message"])
```

```python
messages
```

```python
messages.shape
```

```python
messages['message'].loc[220]
```

```python
#Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
                     
    
```

```python
corpus
```

```python
#Creating a Bag of Words Model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,binary=True,ngram_range=(2,2))
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
X[1]
```

```python
#Label Encoding for spam and ham
y=pd.get_dummies(messages['label'],dtype=int)
y=y.iloc[:,1].values
```

```python
y
```

```python
#Train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train,y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
#prediction
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Now creating a TF-IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X=tf.fit_transform(corpus).toarray()
```

```python
#train-testsplit
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train,y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model1=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model1.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus[0]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
```

```python
model=gensim.models.Word2Vec(words,window=5,min_count=2)
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.wv.similar_by_word('football')
```

```python
model.wv.similar_by_word('money')
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```
