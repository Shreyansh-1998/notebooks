```python
df_sal=pd.read_csv("Salary_Data.csv")
```

```python
df_sal.head()
```

```python
## now checking for multcollinearity
import statsmodels.api as sm
X=df_sal.iloc[:,:-1]
y=df_sal.iloc[:,-1:]
```

```python
## now adding a constant
X=sm.add_constant(X)
```

```python
##fitting ols model
model1=sm.OLS(y,X).fit()
```

```python
model1.summary()
```

Here we can see first that coefficient values are very large, even though r2 is good, then again standard error is very large indicating features being internally correlated. Then to add further more, p-value is very much higher than 0.05.

```python
## To check which features are correlated now
X.iloc[:,1:].corr()
```

We can see age and years of experience are 98% correlated. So we can remove one of them. With the help of p-value, whichever one is more higher drop it.

```python
X.drop(['Age'],axis=1,inplace=True)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
