```python
#Getting independent features
X=df.drop('label',axis=1)
```

```python
#output column
y=df['label']
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
#Defining a vocabulary size
voc_size=10000
```

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
### Preprocessing
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

corpus=[]
for i in range (0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review=review.lower()
    review=review.split()

    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## ONE HOT REPRESENTATION

```python
ohe=[one_hot(word,voc_size)for word in corpus]
ohe
```

## FEATURE REPRESENTATION
This is done to make sure that every feature has the same length and for that we do padding

```python
sentence_length=30
embedded_docs=pad_sequences(ohe,maxlen=sentence_length,padding='pre')
print(embedded_docs)
```

```python
#verifying feature representation
print(embedded_docs[16251])
```

```python
## Creating a model
embedding_vec_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vec_features))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.build(input_shape=(None, sentence_length))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
#train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,random_state=0, test_size=0.2)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
          
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5,1,0) ##AUC ROC Curve
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
