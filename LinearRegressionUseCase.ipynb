```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_cars=pd.read_csv("cars.csv")
```

```python
df_cars.head()
```

```python
df_cars.set_index('Car_ID', inplace=True)
```

```python
df_cars.info()
```

```python
df_cars.describe()
```

```python
df_cars.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
for i in df_cars.select_dtypes('object').columns:
    df_cars[i] = encoder.fit_transform(df_cars[i])

df_cars.head()
```

```python
##Now checking for multi-collinearity
corr=df_cars.iloc[:,1:].corr()
corr.style.background_gradient(cmap='coolwarm')
```

Since none of the features are more than 90% correlated we will move ahead keeping all the columns.

```python
##splitting the dataset
X=df_cars.iloc[:,:-1]
y=df_cars.iloc[:,-1:]
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
y.head()
```

```python
##Doing a train-test split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.20,random_state=0)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
